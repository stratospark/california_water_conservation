September Urban Water Supplier Report
=====================================

Every month, California's State Water Resources Control Board is releasing [Monthly Urban Water Supplier Report Data](http://www.waterboards.ca.gov/waterrights/water_issues/programs/drought/conservation_reporting_info.shtml). Every urban water agency in the state is required to submit residential water consumption information. For the first time, these agencies are also calculating R-GPCD, or residential gallons per-capita per day.



Each month's data is posted as an .xlsx file, which we can read with Pandas:

As you can see, we're going to need to clean up some data.

Let's give the columns names without spaces so that we can use Panda's DataFrame's dot notation, i.e. **df.SupplierName** instead of **df["SupplierName"]**

```python
df.columns = ["SupplierName", "StageInvoked", "MandatoryRestrictions", "ReportingMonth",
              "TotalMonthlyPotable2014", "TotalMonthlyPotable2013", "Units",
              "Qualification", "TotalPopulationServed", "ReportedRGPCD",
              "OptEnforcementActions", "OptImplementation", 
              "OptRecycledWater", "OptRecyledWaterUnits",
              "CalcMonthlyGallons2014", "CalcMonthlyGallons2013",
              "CalcRGPCD2014", "CalcRGPCD2013", "PercentResidentialUse", "Comments"]
df.head()
```

Every supplier reports their monthly consumption using different units.

* G: Gallons
* MG: Millions of Gallons
* AF: Acre-Feet
* CCF: Cubic Feet

We want to standardize on gallons. The report includes a computed column CalcMonthlyGallons2014 and CalcMonthlyGallons2013. We can validate this by applying the appropriate conversion.

```python
df.Units.unique()
```

```python
sorted(df.Qualification.unique())[:10]
```

```python
supplier_means = df.groupby(df.SupplierName).agg({"CalcRGPCD2014": np.mean, "CalcRGPCD2013": np.mean}).sort("CalcRGPCD2014", inplace=False)
supplier_means["Diff"] = (supplier_means["CalcRGPCD2014"]-supplier_means["CalcRGPCD2013"])/supplier_means["CalcRGPCD2013"]*100
supplier_means.sort("Diff")
```

```python
supplier_means.plot(figsize=(15,80), kind="barh")
```

```python
supplier_means["Diff"].sort("Diff", inplace=False).plot(figsize=(15,80), kind="barh")
```
